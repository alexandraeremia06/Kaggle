```python
# Load the data from the first sheet ('Restaurant1') into df_sales_restaurant_1
df_sales_restaurant_1 = pd.read_excel('restaurants_accountability.xlsx', sheet_name='Restaurant1', header=0, parse_dates=['Date'])

# Display the first 5 rows and column names to verify
display(df_sales_restaurant_1.head())
print(df_sales_restaurant_1.columns)
```

Add in each dataframe a new `restaurant_id`column (id = 1 for restaurant_1 and id = 2 for restaurant_2) and concatenate the two dataframes into a single one named `df_sales`.  

Pay attention to the `axis`along which you will concatenate the two dataframes.

```python
# Add a new column 'restaurant_id' with the value 1 to df_sales_restaurant_1
df_sales_restaurant_1['restaurant_id'] = 1

# Add a new column 'restaurant_id' with the value 2 to df_sales_restaurant_2
df_sales_restaurant_2['restaurant_id'] = 2

# Concatenate the two dataframes along the rows (axis=0) into df_sales
df_sales = pd.concat([df_sales_restaurant_1, df_sales_restaurant_2], ignore_index=True)

# Display the first few rows of the combined dataframe and check its info
display(df_sales.head())
display(df_sales.tail())
print(df_sales.info())
```

Export the concatenated dataframe to a new `restaurants_accountability_full.xlsx` file in a sheet called `All`



```python
import pandas as pd

# the concatenated dataframe is named df_sales

# specify the path and filename
output_file = '/content/restaurants_accountability_full.xlsx'

# export to Excel with a sheet named "All"
df_sales.to_excel(output_file, sheet_name='All', index=False, engine='openpyxl')

print(f"DataFrame exported to {output_file}")

# optionally, if I want to download the file from Colab:
from google.colab import files
files.download(output_file)

# update and reorganize the xlsx file
```
