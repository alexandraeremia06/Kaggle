Why does this line of code cause an error?

```python
# df.iloc[1:3, ["B", "C", "D"]]

# display(df.iloc[1:4, [1, 2, 3]])

df.iloc[1:4, [1, 2, 3]]
```

>The error happens because .iloc stands for integer-location based indexing. It means that when you use .iloc, you need to specify rows and columns using their numerical positions, not their labels or names.
Insted, it should be: display(df.iloc[1:4, [1, 2, 3]])   and not: df.iloc[1:3, ["B", "C", "D"]]



## Boolean Indexing


Select rows from `df` where the value in column B is greater than or equal to 8.


```python
df.loc[df["B"] >= 8]

# display(df.loc[df["B"] >= 8])
```

Select rows from df where the value in column B is greater than or equal to 8 and the value in column D is equal to "a".

```python
display(df["B"]>=8, ["D"]=="a")
```

What does this line of code do?

```python
df.loc[df["E"]=="c", "A"] = 8

# df["E"] == "c":
# this part evaluates a condition: for each row in the DataFrame df, it checks if the value in column E is equal to the string 'c'.
# it returns a boolean series (true or false values) where true indicates that the condition is met for that row, and false indicates it's not.
# df.loc[boolean_series, "A"]:
# the .loc accessor in pandas is used for label-based indexing. It takes two main arguments: the rows to select, and the columns to select.
# the boolean_series from the first step is used as the row selector. This means .loc will only consider the rows where df["E"] == "c" evaluated to be true.
# "A" is used as the column selector. This specifies that we are targeting the values in column A for the previously selected rows.
# = 8:
# this is the assignment operator. It takes all the cells identified by the previous two steps (i.e., in column A, but only for the rows where E was 'c') and assigns the integer value 8 to them.
# this line of code finds all rows where column E has the value 'c', and for those specific rows, it changes the value in their corresponding A column to 8.
```
