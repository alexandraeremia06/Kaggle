Hypothesis 1:
Null Hypothesis (H0): There is no difference in customer satisfaction between orders delivered on time and orders delivered late.
Alternative Hypothesis (H1): Orders delivered on time have higher customer satisfaction compared to orders delivered late.

A/B Testing Process:
Divide the dataset into two groups: one group with orders delivered before or on the estimated delivery time and the other with orders delivered after the estimated delivery time.
Compute the average satisfaction rating for each group.
Conduct a hypothesis test (e.g., t-test) to determine if there is a statistically significant difference in satisfaction ratings between the two groups.

Interpretation:
If the p-value is less than a chosen significance level (e.g., 0.05), reject the null hypothesis and conclude that there is a significant difference in satisfaction ratings between on time and late deliveries.
If the p-value is greater than the significance level, fail to reject the null hypothesis, indicating no significant difference in satisfaction ratings between on time and late deliveries.

```python
from scipy.stats import ttest_ind

# t test
# We compare the 'review_score' of on-time deliveries with late deliveries.
# .dropna() is used to handle any missing review scores.
t_stat, p_value = ttest_ind(on_time['review_score'].dropna(), late['review_score'].dropna())
print(f'T-statistic: {t_stat}')
print(f'P-value: {p_value}')

# Interpretation of the results:
# The t-test aims to assess if the mean customer satisfaction (review_score) differs significantly between the two groups (on-time vs. late deliveries).
# - T-statistic: This value measures the size of the difference between the means of the two groups relative to the variation within the groups. A larger absolute T-statistic suggests a greater difference between the group means.
# - P-value: This is the probability of observing a difference as extreme as, or more extreme than, the one calculated, assuming that there is no actual difference between the groups (the null hypothesis).

# As a general rule:
# - If the P-value < 0.05 (a common significance level), we reject the null hypothesis. This means there is a statistically significant difference in customer satisfaction between on-time and late deliveries.
# - If the P-value >= 0.05, we fail to reject the null hypothesis. This suggests there is no statistically significant difference in customer satisfaction between on-time and late deliveries.

# Results:
# A P-value very close to 0.0 observed, indicates a highly significant difference.
# T-statistic (121.496): This value refers to the difference between the average review_score of on-time deliveries and late deliveries,
# relative to the variation within those groups. A large absolute T-statistic (like 121.496) suggests a substantial difference between the mean
# satisfaction scores of the two groups.
# The P-value is 0.0 (extremely small) so we would reject the null hypothesis, concluding that shipping time does affect customer satisfaction.
```

```python
alpha = 0.05

if p_value < alpha:
  print("Reject H0: There is no difference in customer satisfaction between orders delivered on time and orders delivered late.")
else:
  print("Fail to reject H0: here is no difference in customer satisfaction between orders delivered on time and orders delivered late.")
```

Summary of Hypothesis Test Results

Normality Test (Shapiro-Wilk)
We first performed Shapiro-Wilk tests on the `review_score` distributions for both on-time and late deliveries to check for normality. The results showed:
- On-time deliveries: p-value = `8.060000311605506e-146`
- Late deliveries: p-value = `3.961329385975819e-73`

Both p-values are significantly less than the conventional significance level of 0.05. This indicates that the distributions of `review_score` for both groups (on-time and late deliveries) are not equally distributed. This is in contradiction with a key assumption for parametric tests like the t-test, suggesting that a non-parametric alternative might be more appropriate.

Mann-Whitney U Test
Given the non-normal distribution of our data, the non-parametric Mann-Whitney U test (from cell `X3Kc9EGgiPgn`) was conducted to compare customer satisfaction between on-time and late deliveries. The results are:
- U statistic: `530208489.0`
- p-value: `0.0`

With a p-value of 0.0, which is far below the significance level of 0.05, we cannot validate the null hypothesis (H0) that there is no difference in customer satisfaction between orders delivered on time and orders delivered late. The `alternative='greater'` setting implies that on-time deliveries have significantly higher review scores than late deliveries.

Independent Samples T-test
Although the normality assumption was not met, an independent samples t-test was also performed for comparison. The results were:
- T-statistic: `121.4963526457059`
- P-value: `0.0`

Similar to the Mann-Whitney U test, the t-test also yielded a p-value of 0.0, leading to the rejection of the null hypothesis. The large T-statistic indicates a substantial difference in the mean satisfaction scores. The mean satisfaction for on-time deliveries is `4.29` while for late deliveries it is `2.57`.

Conclusion
Based on the Shapiro-Wilk tests, the `review_score` data for both groups (on-time and late deliveries) is not normally distributed. Therefore, the Mann-Whitney U test is the more appropriate statistical test for comparing customer satisfaction between these two groups.

Both the Mann-Whitney U test and the t-test produced a p-value of 0.0. This overwhelmingly strong evidence leads us to reject the null hypothesis (H0: There is no difference in customer satisfaction between orders delivered on time and orders delivered late). Instead, we accept the alternative hypothesis (H1: Orders delivered on time have higher customer satisfaction compared to orders delivered late).

In conclusion, there is a statistically significant difference in customer satisfaction based on shipping time. Specifically, customers who receive their orders on time report significantly higher satisfaction scores (mean: `4.29`) compared to those whose orders are delivered late (mean: `2.57`). This suggests that timely delivery is a critical factor influencing customer satisfaction.
